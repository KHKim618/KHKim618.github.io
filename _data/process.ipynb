````python
pattern = re.compile(r'<li>(?:\s+?)<p>((?:.|\s)*?\.) \((\d{4})\)(?:\s+?)\"((?:.|\s)*?)\"(?:\s+?)((?:.|\s)*?)</p>(?:\s+?)</li>')
````

````python
with open('/Users/songk/Downloads/KIM_LAB_Publications_No_Empty_Tags.html', 'r') as f:
    data = f.read()
````

````python
results = re.findall(pattern, data)
````

````python
results_processed = []
for res in results:
    res_processed = []
    for r in res:
        r = re.sub(r'\s+', ' ', r)
        r = r.strip()
        res_processed.append(r)
    data = {}
    data['author_list'] = res_processed[0]
    data['year'] = res_processed[1]
    data['title'] = res_processed[2]
    data['pub_info'] = res_processed[3]
    results_processed.append(data)
````
